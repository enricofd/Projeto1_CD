## Pergunta:

É possível relacionar um maior PIB per capita de países membros da União Européia com suas balanças comerciais quando comparado com o restante da Europa?

## Explicação da hipótese e mecanismo utilizado:

É esperado que países da União Europeia possuam uma balança comercial mais positiva como sendo consequência do livre comércio nessa. A balança comercial positiva, por sua vez, implicaria em uma maior abundância econômica nesses países e portanto um maior PIB per capita.

Para verificar essa hipótese, foram comparadas as balanças comerciais da União Europeia com a do restante da Europa e o PIB per capita desses mesmos grupos. Com esses resultados geraram-se gráficos que relacionam justamente esses dados para possibilitar uma conclusão.

## Bases de dados usadas:

https://www.gapminder.org/data/

## Análise de dados:

Importando as bibliotecas:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

Importando as tabelas:

```python
Trade_raw = pd.read_excel('Trade balance.xlsx')
GDPperC_raw = pd.read_excel('GDPpercapitaconstant2000US.xlsx')
```

Colocando Países como index:

```python
Trade_indexed = Trade_raw.set_index('Trade balance (% of GDP)')
GDPperC_indexed = GDPperC_raw.set_index('Income per person (fixed 2000 US$)')
```

Selecionando os países da Europa:

```python
Europe_members = ['Andorra',  'Austria', 'Belgium', 'Bulgaria', 'Bosnia and Herzegovina',  'Channel Islands',  'Cyprus', 'Germany',
 'Denmark', 'Spain',  'Estonia', 'Finland',  'France', 'United Kingdom',  'Gibraltar',  'Greece', 'Croatia', 'Hungary',
 'Isle of Man', 'Ireland',  'Iceland',  'Italy',  'Liechtenstein',  'Lithuania', 'Latvia',  'Monaco',  'Moldova', 
 'Malta',  'Montenegro', 'Netherlands',  'Poland',  'Portugal',  'Romania',  'Slovak Republic',  'Slovenia',  'Sweden',  'Turkey',
 'Kosovo']
```

Selecionando os países membros da União Europeia no ano de 2000:

```python
UE_participants_2000 = ["Belgium", "France", "Italy", "Germany", "Luxembourg", "Netherlands", 'United Kingdom', "Denmark", "Ireland", "Greece", "Portugal", "Spain","Austria", "Finland", "Sweden"]
```

Analisando a balança comercial mundial entre os anos de 1970 e 2010:

```python
Trade_years = Trade_indexed.loc[:,('1970', '1971', '1972', '1973', '1974', '1975', '1976', '1977',
       '1978', '1979', '1980', '1981', '1982', '1983', '1984', '1985', '1986',
       '1987', '1988', '1989', '1990', '1991', '1992', '1993', '1994', '1995',
       '1996', '1997', '1998', '1999', '2000', '2001', '2002', '2003', '2004',
       '2005', '2006', '2007', '2008', '2009', '2010')]

Trade_T = Trade_years.T

medias_Trade = Trade_years.mean()
plt.figure(figsize=(30,10))
for Country in Trade_T.columns:
    plt.scatter(x=Trade_T.index, y=Trade_T[Country])
plt.title(' Balança comercial em % do PIB X Tempo ')
plt.ylabel('Balança comercial em % do PIB')
plt.xlabel('Tempo em anos')
plt.plot(medias_Trade.index,  medias_Trade, color="r");
```

Interpretando o gráfico, pode-se analisar que a balança comercial média ao longo do tempo se manteve praticamente contstante, ao redor de 0% do PIB mundial. Pode-se compreender, portanto, que não houveram anos em que uma balança comercial positiva para dado grupo seria justificada por um aumento global.

Comparando as balanças comerciais dos países da União Europeia com os do restante da Europa para o ano de 2005:

```python
Trade_indexed.loc[:,"Europe"] = "No"
Trade_indexed.loc[:,"UE_Members"] = "No"
Trade_indexed.loc[Europe_members, "Europe"] = "Yes"
Trade_indexed.loc[UE_participants_2000, "UE_Members"] = "Yes"
Trade_Europe = Trade_indexed.loc[Trade_indexed.Europe == "Yes"]
Trade_Europe.boxplot(column="2005", by="UE_Members");
plt.title('Balança comercial em % do PIB X Mebro da União Europeia');
plt.ylabel('Balança comercial em % do PIB em 2005');
plt.xlabel('Membro da União Europeia');
plt.suptitle("");
Trade_Europe.hist(column="2005", by="UE_Members", density=True);
plt.ylabel('Quantidade de países em porcentagem');
plt.xlabel('Balança comercial em % do PIB em 2005');
plt.suptitle("");
```

Por meio dessa análise pode se concluir que para o ano de 2005 a União Europeia quando comparado com o restante da Europa possuia uma balança comercial em termos de PIB mais positiva.

Comparando as balanças comerciais dos países da União Europeia com os do restante da Europa para o ano de 2010 para assegurar que o resultado anterior não foi um evento único:

```python
Trade_Europe.boxplot(column="2010", by="UE_Members");
plt.title('Balança comercial em % do PIB X Mebro da União Europeia');
plt.ylabel('Balança comercial em % do PIB em 2010');
plt.xlabel('Membro da União Europeia');
plt.suptitle("");
Trade_Europe.hist(column="2010", by="UE_Members", density=True);
plt.ylabel('Quantidade de países em porcentagem');
plt.xlabel('Balança comercial em % do PIB em 2010');
plt.suptitle("");
```

Como se pode notar, a diferença entre as médias das balanças comercias decresceu, entretanto o mesmo padrão continua presente, ou seja, comprova-se o resultado prévio .

Analisando o PIB per capita mundial entre os anos de 1970 e 2010:

```python
GDPperC_years = GDPperC_indexed.loc[:,('1970', '1971', '1972', '1973', '1974', '1975', '1976', '1977',
       '1978', '1979', '1980', '1981', '1982', '1983', '1984', '1985', '1986',
       '1987', '1988', '1989', '1990', '1991', '1992', '1993', '1994', '1995',
       '1996', '1997', '1998', '1999', '2000', '2001', '2002', '2003', '2004',
       '2005', '2006', '2007', '2008', '2009', '2010')]

GDPperC_T = GDPperC_years.T

medias_GDPperC = GDPperC_years.mean()
plt.figure(figsize=(30,10))
for pais in GDPperC_T.columns:
    plt.scatter(x=GDPperC_T.index, y=GDPperC_T[pais])
plt.title(' Pib per capita X Tempo ')
plt.ylabel('Pib per capita em US$')
plt.xlabel('Tempo em anos')
plt.plot(medias_GDPperC.index,  medias_GDPperC, color="r");
```

Interpretando o gráfico, pode-se analisar que o PIB per capita médio ao longo do tempo se manteve praticamente contstante. Pode-se compreender, portanto, que não houveram anos em que um PIB per capita positivo para dado grupo seria justificada por um aumento global.

Comparando os PIB per capita dos países da União Europeia com os do restante da Europa para o ano de 2005:

```python
GDPperC_indexed.loc[:,"Europe"] = "No"
GDPperC_indexed.loc[:,"UE_Members"] = "No"
GDPperC_indexed.loc[Europe_members, "Europe"] = "Yes"
GDPperC_indexed.loc[UE_participants_2000, "UE_Members"] = "Yes"
GDPperC_Europe = GDPperC_indexed.loc[GDPperC_indexed.Europe == "Yes"]
GDPperC_Europe.boxplot(column="2005", by="UE_Members");
plt.title('PIB per capita X Mebro da União Europeia');
plt.ylabel('PIB per capita em 2005');
plt.xlabel('Membro da União Europeia');
plt.suptitle("");
GDPperC_Europe.hist(column="2005", by="UE_Members", density=True);
plt.ylabel('Quantidade de países em porcentagem');
plt.xlabel('PIB per capita em 2005');
plt.suptitle("");
```

Por meio dessa análise pode se concluir que para o ano de 2005 a União Europeia quando comparado com o restante da Europa possuia um PIB per capita mais elevado.

Comparando o PIB per capita dos países da União Europeia com os do restante da Europa para o ano de 2010 para assegurar que o resultado anterior não foi um evento único:

```python
GDPperC_Europe.boxplot(column="2010", by="UE_Members");
plt.title('PIB per capita X Mebro da União Europeia');
plt.ylabel('PIB per capita em 2010');
plt.xlabel('Membro da União Europeia');
plt.suptitle("");
GDPperC_Europe.hist(column="2010", by="UE_Members", density=True);
plt.ylabel('Quantidade de países em porcentagem');
plt.xlabel('PIB per capita em 2010');
plt.suptitle("");
```

Como se pode notar, a diferença entre os PIB per capita aumentou, ou seja, assegura o resultado anterior da disparidade evidenciada.

Comparando a relação entre PIB per capita e balança comercial para os dois grupos analisados nos anos de 2005 e 2010:

Para o ano de 2005:

```python
lista1 = []

def reta1(dados_x, dados_y, c):    
    a = dados_y.cov(dados_x) / dados_x.var()
    b = dados_y.mean() - a*dados_x.mean()
    lista1.append(a)
    plt.scatter(dados_x, dados_y, c=c, alpha=0.8)
    plt.title('PIB per capita em USD  X Balança comercial em % do PIB ')
    plt.ylabel('PIB per capita em USD em 2005')
    plt.xlabel('Balança comercial em % do PIB em 2005')

    plt.plot((dados_x.min(), dados_x.max()), (a*dados_x.min()+b, a*dados_x.max()+b), color=c)

    plt.tight_layout()
    
plt.scatter(Trade_Europe.loc[Trade_Europe["UE_Members"] == "Yes","2005"], GDPperC_Europe.loc[GDPperC_Europe["UE_Members"] == "Yes", "2005"], color ="r")
reta1(Trade_Europe.loc[Trade_Europe["UE_Members"] == "Yes","2005"],  GDPperC_Europe.loc[ GDPperC_Europe["UE_Members"] == "Yes", "2005"], "r")
plt.scatter(Trade_Europe.loc[Trade_Europe["UE_Members"] == "No","2005"], GDPperC_Europe.loc[GDPperC_Europe["UE_Members"] == "No", "2005"], color = "b")
reta1(Trade_Europe.loc[Trade_Europe["UE_Members"] == "No","2005"],  GDPperC_Europe.loc[ GDPperC_Europe["UE_Members"] == "No", "2005"], "b");
print ("Pontos em vermelho = Países membros da União Europeia")
print ("Pontos em azul = Países do resto da Europa")
```

Por meio desse gráfico poode ser estabelecida para o ano de 2005 uma proporção entre a balança comercial e PIB per capita de forma que essa é expressa pelos coeficientes angulares das retas plotadas. Ou seja para a reta vermelha, representante dos países membros da União Europeia, pode se notar um coeficiente maior do que o da reta azul, que representa o restante dos países da Europa:

```python
print(f'O coeficiente angular da reta representante da União Europeia vale {lista1[0]}.')
print(f'O coeficiente angular da reta representante dos países europeus que não fazem parte da União Europeia vale {lista1[1]}.')
print(f'Portanto, o coeficiente angular da reta da União Europeia é {lista1[0]/lista1[1]} vezes maior que o coeficiente angular da reta dos não membros dessa.')
```

Desta forma é possível estabelecer para esse gráfico que quanto maior a balança comercial, maior o PIB per capita, ou seja, comprova-se a hipótese inicial.

Para 2010:

```python
lista2 = []

def reta2(dados_x, dados_y, c):    
    a = dados_y.cov(dados_x) / dados_x.var()
    b = dados_y.mean() - a*dados_x.mean()
    lista2.append(a)
    plt.scatter(dados_x, dados_y, c=c, alpha=0.8)
    plt.title('PIB per capita em USD  X Balança comercial em % do PIB ')
    plt.ylabel('PIB per capita em USD em 2010')
    plt.xlabel('Balança comercial em % do PIB em 2010')

    plt.plot((dados_x.min(), dados_x.max()), (a*dados_x.min()+b, a*dados_x.max()+b), color=c)

    plt.tight_layout()

plt.scatter(Trade_Europe.loc[Trade_Europe["UE_Members"] == "Yes","2010"], GDPperC_Europe.loc[GDPperC_Europe["UE_Members"] == "Yes", "2010"], color ="r")
reta2(Trade_Europe.loc[Trade_Europe["UE_Members"] == "Yes","2010"],  GDPperC_Europe.loc[ GDPperC_Europe["UE_Members"] == "Yes", "2010"], "r")
plt.scatter(Trade_Europe.loc[Trade_Europe["UE_Members"] == "No","2010"], GDPperC_Europe.loc[GDPperC_Europe["UE_Members"] == "No", "2010"], color = "b")
reta2(Trade_Europe.loc[Trade_Europe["UE_Members"] == "No","2010"],  GDPperC_Europe.loc[ GDPperC_Europe["UE_Members"] == "No", "2010"], "b");
print ("Pontos em vermelho = Países membros da União Europeia")
print ("Pontos em azul = Países do resto da Europa")
```

Para esse gráfico pode-se também notar um maior coeficiente angular da reta vermelha, representante dos Membros da União Europeia:

```python
print(f'O coeficiente angular da reta representante da União Europeia vale {lista2[0]}.')
print(f'O coeficiente angular da reta representante dos países europeus que não fazem parte da União Europeia vale {lista2[1]}.')
print(f'Portanto, o coeficiente angular da reta da União Europeia é {lista2[0]/lista2[1]} vezes maior que o coeficiente angular da reta dos não membros dessa.')
```

O que vale ser notado, entretanto é o aumento ou melhora da balança comercial dos países não membros da União Europeia e como isso impactou o aumento do PIB per capita. Dessa forma, o coeficiente da respectiva reta também aumentou. Os membros da União Européia, por sua vez, para esse mesmo ano, tiveram uma melhora em sua balança comercial, como pode ser visto a seguir, e possuiram também um aumento na média do PIB per capita, mesm que o coeficiente angular da reta teha diminuido. 

```python
MT_UE = Trade_Europe[Trade_Europe['UE_Members'] == "Yes"].mean()
MPpC_UE = GDPperC_Europe[GDPperC_Europe['UE_Members'] == "Yes"].mean()
VT = MT_UE['2010'] - MT_UE['2005']
VPpC = MPpC_UE['2010'] - MPpC_UE['2005']

data = {'2005':[MT_UE['2005'], MPpC_UE['2005']], '2010':[MT_UE['2010'], MPpC_UE['2010']], 'Variação entre 2005 e 2010':[VT,VPpC]}
df = pd.DataFrame(data, index =['Balança comercial em % do PIB', 'PIB per Capita'])
df
```

## Conclusão:

Por meio dos dados analisados pode-se chegar a conclusão de que não é plausível relacionar o aumento da balança comercial com o aumento do PIB per capita. Esse fator pode muito bem ser compreendido pela interpretação do último gráfico de dispersão plotado:

```python
plt.scatter(Trade_Europe.loc[Trade_Europe["UE_Members"] == "Yes","2010"], GDPperC_Europe.loc[GDPperC_Europe["UE_Members"] == "Yes", "2010"], color ="r")
reta2(Trade_Europe.loc[Trade_Europe["UE_Members"] == "Yes","2010"],  GDPperC_Europe.loc[ GDPperC_Europe["UE_Members"] == "Yes", "2010"], "r")
plt.scatter(Trade_Europe.loc[Trade_Europe["UE_Members"] == "No","2010"], GDPperC_Europe.loc[GDPperC_Europe["UE_Members"] == "No", "2010"], color = "b")
reta2(Trade_Europe.loc[Trade_Europe["UE_Members"] == "No","2010"],  GDPperC_Europe.loc[ GDPperC_Europe["UE_Members"] == "No", "2010"], "b");
print ("Pontos em vermelho = Países membros da União Europeia")
print ("Pontos em azul = Países do resto da Europa");
```

Pode-se notar um ponto vermelho na extema direita, um outliner, que é responsável por fazer com que as média calculadas na última tabela sejam maiores do que realmente são. Esse fator é muito bem expresso pelo coeficiente angular da reta vermelha, que em comparação com a ano de 2005 decresceu.

Portanto, em suma, assim como o artigo "Does free trade result in higher GDP per capita? An international perspective", escrito por John Lambrechts, Erin McGrath, Naomi Rule em 2012 demonstra, não é possivel estabelecer uma relação diretamente proporcional entre aumento da balança comercial e o aumento do PIB per capita, o perceptivel muitas vezes é o oposto, uma inversão da proporcionalidade. Desta forma, o fato da União Européia possuir um PIB per capita maior que o restante da Europa não é justificado por sua balança comercial.
